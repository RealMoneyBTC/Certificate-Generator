```python
# ✅ Step 4: User Input
name = input("Enter student name: ")
course = input("Enter course name: ")
date = input("Enter date (e.g., 2025-07-01): ")
signing_key = input("Enter your GPG key ID (e.g., CBC785...): ")
```
